```python
import numpy as np
import pylab as p
import matplotlib.pyplot  as  plt
import matplotlib.cm as cm

import math
import itertools 

np.set_printoptions(linewidth = 130, precision = 13)
%matplotlib inline

from scipy.optimize import fsolve
from pylab import meshgrid,cm,imshow,contour,clabel,colorbar,axis,title,show
from matplotlib import colors
from matplotlib import pyplot
```

```python
def resistance(beta, del_b):
    R0 = beta/(gamm + r)
    mu = 1/(k*(R0 - 1))
    theta2 = R0/2*mu*(2*mu + 1)/(mu + 1)**2
    theta4 = beta/2*(R0 - 1)*mu/(mu + 1)**2
    
    #lambda^2_1 + lambda^1_2
    summ = theta4*(alp[0,1] + alp[1,0] - alp[0,0] - alp[1,1])
    lambd12 = -del_b - theta2*del_gamm + theta4*(mu*(alp[0][1] - alp[1][0]) + alp[0][1] - alp[0][0])
    lambd21 = summ - lambd12
    
    return lambd21, lambd12, lambd12/summ
```

```python
r = 0.3
alp = np.array([[0, np.sqrt(2)], [np.sqrt(2), 0]])
gamm = 1.2
k = 1

del_gamm = 1
X = np.linspace(1.5, 9, 400)
Y = np.linspace(-10, 10, 1000)

x, y = np.meshgrid(X, Y)
L21, L12, z = resistance(x, y)
pos1 = L21 > 0
pos2 = L12 > 0
neg1 = L21 < 0
neg2 = L12 < 0

coex = pos1*pos2
s2 =   1.1*pos2*neg1
s1 = (-0.1)*pos1*neg2


plt.figure(figsize=(5, 4))
fig1 = plt.contour(x, y, z*coex + s1 + s2, levels = [0, 1], colors=('k',),linestyles=('-',),linewidths=(1,))
plt.contourf(x, y, z*coex + s1 + s2)


plt.ylabel('Variation in transmission rate $\Delta \\beta$', fontsize = 14)
plt.xlabel('Mean transmission rate $\\beta$', fontsize = 14)
plt.colorbar()
plt.savefig('4.2a.jpg')
```
