```python
def resistance(gamm):
    R0 = beta/(gamm + r)
    mu = 1/(k*(R0 - 1))
    theta2 = (R0/2)*(mu*(2*mu + 1)/(mu + 1)**2)
    theta4 = beta/(2*k)*1/(mu + 1)**2
    
    #lambda^2_1 + lambda^1_2
    summ = theta4*(alp[0,1] + alp[1,0] - alp[0,0] - alp[1,1])
    lambd12 = -del_b - theta2*del_gamm + theta4*(mu*(alp[0][1] - alp[1][0]) + alp[0][1] - alp[0][0])
    lambd21 = summ - lambd12
    
    return lambd12/summ
```

```python
r = 0.3
alp = np.array([[0, np.sqrt(2)], [np.sqrt(2), 0]])
k = 1
beta = 6

fig1, ax = plt.subplots(1, 2, sharey='row', figsize=(12,4))
fig1.subplots_adjust(wspace = 1)

delnu_arr = np.array([-1, 0.5, 1])
delb_arr = np.array([0.5, 1])
ls = ['-', '--', ':']

gamm_arr = np.linspace(1.2, 5.7, 1000)

for i in range (0, 2):
    for j in range (0, 3):
        del_b = delb_arr[i]
        del_gamm = delnu_arr[j]
        
        res = []
        
        for gamm in gamm_arr:
            z2 = resistance(gamm)
            res.append(z2)
        
        ax[i].plot(gamm_arr, res, 'k', linestyle = ls[j], label='$\Delta \\nu =$ {}'.format(del_gamm))
    ax[i].set_xlim([1.2, 5.7])
    ax[i].set_ylim([0, 1])
    ax[i].legend(loc="best")
    ax[i].set_xlabel('Mean clearance rate $\gamma$', fontsize = 14)
    ax[i].set_ylabel('Equilibrium of strain-2 prevalence $z^*_2$', fontsize = 14)

ax[0].text(-0.1, 1.1, 'a', transform=ax[0].transAxes, size=16, weight='bold')
ax[1].text(-0.1, 1.1, 'b', transform=ax[1].transAxes, size=16, weight='bold')
    
#fig1.text(0.5, 0.04, 'Mean clearance rate $\gamma$', fontsize = 14, ha='center')
#fig1.text(0.04, 0.5, 'Frequency of strain 2', fontsize = 14, va='center', rotation='vertical')

cols = ['$\Delta b =$ {}'.format(del_b) for del_b in delb_arr]

for ax, col in zip(ax, cols):
    ax.set_title(col)
    
fig1.savefig('5.1.jpg')
```
